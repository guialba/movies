```python
import pandas as pd
```

```python
df = pd.read_csv('movies_scores.csv')
df.drop(['Trailer'], axis=1, inplace=True)
df.head()
```

```python
from api_imdb.movies import MovieSearcher 

def search(row):
    try:
        api = MovieSearcher(row[0])
        return pd.Series(api.data)
    except:
        print("Erro: {}".format(row[0]))


dfMovies = df.apply(search, axis=1)
```
